# 4. Analyzing the results

## Utilities

***The idea now would be to look for the best models (in MSE sense) for each combination (so the mean MSE across the 5 folds for the same analyzed combination), which theoretically should be the features that have the most predictive value***.

Now, we define a function to retrieve the n best-performing subsets per fold, where the input is a dataframe of results such as the ones above:

```python
def top_n_per_fold(df: pd.DataFrame, source_combinations: list[str], n: int):

    """
    Retrieves the rows from the original DataFrame corresponding to the top n combinations
    with the highest mean Negative MSE.

    Parameters:
    - df (pd.DataFrame): The original DataFrame containing 'Negative MSE' and combination columns.
    - source_combinations (list of str): List of column names used to identify combinations.
    - n (int): The number of top combinations to retrieve based on the mean Negative MSE.

    Returns 3 pd.DataFrame:
    - A DataFrame containing the rows with the top N combinations by mean Negative MSE.
    - A DataFrame containing the unique combination IDs and their mean Negative MSE
    - A DataFrame with binary inclusion indicators for features in the n top combinations.
    """

    # Compute the mean MSE across folds
    df_mean_MSE = df.groupby(source_combinations)['Negative MSE'].mean().reset_index()

    # Retrieve those n observations with the highest negative MSE (i.e., lowest MSE)
    df_top_n_combinations_mean_MSE = df_mean_MSE.nlargest(n, 'Negative MSE')

    # Retrieve the ids of those observations
    df_top_n_combinations_mean_MSE_ids = df_top_n_combinations_mean_MSE[source_combinations]

    # Merge with the original DataFrame to get all rows corresponding to the top n combinations
    df_top_observations = df.merge(df_top_n_combinations_mean_MSE_ids, on = source_combinations, how = 'inner')

    # Identify feature columns
    feature_columns = [
        col for col in df.columns if col not in source_combinations + ['Fold', 'Negative MSE', 'Alpha', 'Group_reg_parameter', 'Intercept']
        ]
    
    # Create a list to hold the inclusion status for each feature in each combination
    inclusion_status_list = []

    # We iterate over the values of the rows (ignoring the index)
    for _, row in df_top_observations.iterrows():
        # Create a dictionary to hold inclusion status for each feature, where
        # we don't include coefficients with NaN or numerically insignificant
        inclusion_status = {col: 1 if (not pd.isna(row[col]) and round(abs(row[col]), 4) > 0) else 0 for col in feature_columns}
        inclusion_status_list.append(inclusion_status)

    # Convert the inclusion status list to a DataFrame
    inclusion_df = pd.DataFrame(inclusion_status_list)

    # Returns the data frame with the top results, the mean MSE of the top combinations, and the inclusion DataFrame
    return df_top_observations, df_top_n_combinations_mean_MSE, inclusion_df
```

## Overall conclusions
We can start by creating a data frame which compares the inclusion of features in the different results of datasets and functions.

```python
# We create a data frame with the different number of inclusions
overall_inclusion = pd.DataFrame(data = total_inclusions_v1_f1, columns = ['num_inclusions_v1_f1'])
overall_inclusion = overall_inclusion.join(pd.DataFrame(total_inclusions_v1_f3, columns = ['num_inclusions_v1_f3']))
overall_inclusion = overall_inclusion.join(pd.DataFrame(total_inclusions_v3_f4, columns = ['num_inclusions_v3_f4']))

# We add a column with the mean number of inclusions
overall_inclusion['mean_n_inclusions'] = round((overall_inclusion['num_inclusions_v1_f1'] + overall_inclusion['num_inclusions_v1_f3'] + overall_inclusion['num_inclusions_v3_f4']) / 3, 2)

overall_inclusion
```

Take into account that the maximum number of inclusions is the number of top combinations selected times the number of folds (5 folds). So if 3 combinations have been selected, the max number of inclusions is 3 * 5 = 15.

```python
# We can sort the data frame by the mean number of inclusions
overall_inclusion.sort_values('mean_n_inclusions', ascending = False)
```
